```python
import pandas as pd
```

```python
df1 = pd.read_excel('First Call Solution (Solar).xlsx', sheet_name='Sheet1')
df1.head()
```

```python
df1.shape
```

```python
df1 = df1.drop(['N/A', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14'], axis=1)
df1.shape
```

```python
df1.head()
```

```python
df1 = df1.rename(columns={
    'Patricia': 'First',
    'Rohena': 'Last',
    3212178631: 'PH. NUMBER',
    '2723 Muscatello St': 'Address',
    'Orlando,': 'City',
    'FL': 'State',
    32837: 'Zip'
})
df1.head()
```

```python
df1.info()
```

```python
df1.dropna(axis=0, inplace=True)
df1.isnull().sum()
```

```python
df1['NAME'] = df1['First'].astype(str) + ' ' + df1['Last'].astype(str)
df1.head()
```

```python
df1['ADDRESS'] = df1['Address'].astype(str) + ' ' + df1['City'].astype(str) + ' ' + df1['State'].astype(str) + ' ' + df1['Zip'].astype(str)
df1.head()
```

```python
df1.drop(['First', 'Last', 'Address', 'City', 'State', 'Zip'], axis=1, inplace=True)
df1.head()
```

```python
df1.shape
```

```python
df1.isnull().sum()
```
